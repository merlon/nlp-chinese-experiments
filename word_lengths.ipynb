# Word lengths in chinese

What is the length distribution of chinese words in this annotated dataset https://www.microsoft.com/en-us/download/details.aspx?id=52531 ?

```python
%matplotlib notebook
from Dataset import parse_dataset
from collections import defaultdict
```

```python
sentences=parse_dataset('msra-chinese-word-segmentation-data-v1/msra_bakeoff3_training.xml')
```

```python
count=defaultdict(int)
for s in sentences:
    for w in s:
        count[len(w)]+=1
```

```python
import numpy as np
x,y=zip(*sorted(count.items(),key=lambda x:x[0]))
norm=sum(y)
y_norm=[i/norm for i in y]
y_cum=np.cumsum(y_norm)
```

```python
y_cum[:10]
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y_norm,'.',label="length distribution")
plt.plot(x,np.cumsum(y_norm),'+',label='cumulative length distribution')
plt.legend()
```

# Conclusion
95 % words in text are three characters or less, 90 % is 2 characters or less
